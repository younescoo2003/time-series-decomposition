# Cell 1: Import

```python
import sys
sys.path.append('..')
import pandas as pd
import matplotlib.pyplot as plt
from src.data_loader import load_air_passengers
from src.decomposer import decompose_additive, decompose_multiplicative
from src.visualizer import plot_decomposition, plot_seasonal_pattern
```

# Cell 2: Load Data

```python
data = load_air_passengers()
data.head()
```

# Cell 3: Plot Original

```python
plt.figure(figsize=(12,4))
plt.plot(data)
plt.title('Air Passengers')
plt.show()
```

# Cell 4: Additive Decomposition

```python
result_add = decompose_additive(data['value'])
plot_decomposition(data['value'], result_add, 'Additive Decomposition')
plt.savefig('../results/plots/additive_decomposition.png')
```

# Cell 5: Multiplicative Decomposition

```python
result_mul = decompose_multiplicative(data['value'])
plot_decomposition(data['value'], result_mul, 'Multiplicative Decomposition')
plt.savefig('../results/plots/multiplicative_decomposition.png')
```

# Cell 6: Seasonal Pattern

```python
plot_seasonal_pattern(result_add)
plt.savefig('../results/plots/seasonal_pattern.png')
```
